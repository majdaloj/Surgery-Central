```python
with open('hospital_contact.txt', encoding="utf-8") as f:
    lines = f.readlines()
```

```python
lines = [line.strip() for line in lines]
```

```python
with open('hospitals.txt') as f:
    lines1 = f.readlines()
    lines1 = [line.strip() for line in lines1]
```

```python
lines1
```

```python
hospital_dic = {}

for line in lines1:
    hospital_dic[line] = {}
```

```python
hospital_dic
```

```python
current_key = "Unity Health Toronto - St. Michael's"

for line in lines:
    if line in list(hospital_dic.keys()):
        current_key = line
    
    if line not in list(hospital_dic.keys()) and len(line) > 0:
        key = line[0:line.index(':')]
        value = line[line.index(':')+2:]
        
        hospital_dic[current_key][key] = value
```
